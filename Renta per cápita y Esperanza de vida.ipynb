## 11. Bonus: Evaluamos la regresión lineal expuesta arriba

#### Una regresión lineal trata de predecir la variable y de una observación y mediante la variable x de dicha observación:
$y_i = \alpha + {\beta}x_i + \epsilon_i$

#### donde se intenta encontrar la constante y la pendiente de la recta que minimicen el error (residuo) de todas las observaciones al cuadrado:

$SSR = \sum \limits _{i=1} ^{n}(y_i - (\alpha + {\beta}x_i))^2$

#### Ya que vamos a realizar una regresión ponderada, debemos añadir un peso a la hora de minimzar la función de perdida:

$SSR_w = \sum \limits _{i=1} ^{n}(w_iy_i - w_i(\alpha + {\beta}x_i))^2$

```python
# importamos statsmodel (se parece más a R y creo que es mejor para realizar regresiones)
import statsmodels.api as sm

# Definimos variables y añadimos la constante

Y = df["Esperanza de vida"]
X = sm.add_constant(df["log_renta_pc"])
W = df["Poblacion"]

# modelo de regresión lineal con residuos ponderados por población
model = sm.WLS(Y, X, weights = W).fit()
model.summary()
```

### 11.1. Estimamos predicción y residuos al cuadrado de cada observación

```python
df["Prediction"] = model.predict(X)

df["Residuals"] = abs(df["Residuals"])
       
df.tail()
```

### 11.2. Comprobamos supuestos

#### Los residuos deben seguir aproximadamente una normal de media cero y desviación típica sigma
$\epsilon_i = y_i - (\alpha + {\beta}x_i)$

$\epsilon \sim \mathcal{N}(0, \sigma)$

```python
sm.qqplot(model.resid, line ='45')
plt.title("El modelo está muy lejos de la normalidad, pese a la transformación logarítimica");
```

#### La homocesticidad implica que la varianza de los residuos es constante 
$Var(\epsilon) = \sigma^2$ $\forall$  $i...n$

```python
plt.scatter(model.predict(X), model.resid)
plt.axhline(np.mean(model.resid), color="red")
plt.title("Los residuos parecen distribuirse aleatoriamente")
plt.xlabel("predicciones")
plt.ylabel("residuos");
```

## 12. Bonus: ¿Qué países están más lejos del modelo?
